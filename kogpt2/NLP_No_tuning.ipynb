```python
# 2) PyTorch, HuggingFace Transformers 임포트 및 디바이스 설정
import torch
from transformers import PreTrainedTokenizerFast, GPT2LMHeadModel

# GPU 사용 가능하면 GPU로, 아니면 CPU로
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# 3) KoGPT2 토크나이저 & 모델 불러오기
# 모델 이름: "skt/kogpt2-base-v2" (SKT에서 공개한 KoGPT2)
tokenizer = PreTrainedTokenizerFast.from_pretrained(
    "skt/kogpt2-base-v2",
    bos_token='</s>', eos_token='</s>',
    unk_token='<unk>', pad_token='<pad>', mask_token='<mask>'
)
model = GPT2LMHeadModel.from_pretrained("skt/kogpt2-base-v2").to(device)
model.eval()  # 생성 모드로 변경 (dropout off)
```

```python
# 4) 간단한 텍스트 생성 함수 정의
def generate_korean_text(
    prompt,         # 문자열: 모델 입력 텍스트
    max_length=500,  # int: 생성할 최대 토큰 길이
    top_k=50,       # int: top-k sampling 개수
    top_p=0.95,     # float: nucleus sampling 확률 임계치
    repetition_penalty=1.2   # float: 반복 패널티
):
    """
    prompt: 생성 시작 문자열 (예: "안녕하세요. ")
    반환값: 생성된 문자열 (prompt + continuation)
    """
    # 1) 입력 문장 토큰화
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    # 2) generate() 호출
    output_ids = model.generate(
        input_ids,
        max_length=max_length,
        num_beams=1,             # 빔 서치 사용 안 함
        do_sample=True,          # 샘플링 모드
        top_k=top_k,
        top_p=top_p,
        repetition_penalty=repetition_penalty,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id
    )

    # 3) 토큰 시퀀스를 문자열로 디코딩
    generated = tokenizer.decode(output_ids[0].tolist(), skip_special_tokens=True)
    return generated
```

```python
# 5) 실제로 몇 가지 예시를 입력해 보고 결과 출력해 보기
if __name__ == "__main__":
    prompts = [
        "사랑에 관한 시:",
        "사랑에 관한 시: 이제 한평생 내 가슴에",
        "<|topic: 사랑|>"
    ]

    for p in prompts:
        print("▶ Prompt:", p)
        print("▶ Output:", generate_korean_text(p, max_length=60))
        print('\n' + "-" * 50 + '\n')
```
