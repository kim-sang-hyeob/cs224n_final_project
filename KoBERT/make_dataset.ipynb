# 한국어 감정 정보 연속적 대화 데이터셋

```python
import pandas as pd
import numpy as np
dlg = pd.read_excel('한국어 감정 정보가 포함된 연속적 대화 데이터셋\한국어_연속적_대화_데이터셋.xlsx') 
dlg.head()
```

```python
new_columns=dlg.iloc[0]
dlg_final=dlg.rename(columns=new_columns).iloc[:,1:3]
```

```python
dlg_final["감정"].value_counts()
```

```python
dlg_final=dlg_final[dlg_final["감정"].isin(["중립","놀람","분노","슬픔","행복","혐오","공포"])]
```

```python
korean_to_final = {
    "공포": "불안",
    "놀람": "당황",
    "분노": "분노",
    "슬픔": "슬픔",
    "중립": "무감정",
    "행복": "기쁨",
    "혐오": "상처"
}

if "감정" not in dlg_final.columns:
    raise KeyError(" no error here ")
```

```python
dlg_final
```

```python
dlg_final["최종라벨"] = dlg_final["감정"].map(korean_to_final)
```

```python
desired_counts = {
    "불안":    98,
    "당황":   100,
    "분노":   100,
    "슬픔":   700,
    "무감정": 10000,
    "기쁨":  1030,
    "상처":   220
}
```

```python
available_counts = dlg_final["최종라벨"].value_counts().to_dict()
print("▶ 세 번째 데이터풀 가용 개수:", available_counts)
```

```python
for label, needed in desired_counts.items():
    avail = available_counts.get(label, 0)
    if avail < needed:
        raise ValueError(f"🔴 라벨 '{label}'는 풀에 {avail}개만 있는데, {needed}개를 뽑으려고 시도했습니다.")
```

```python
sampled_list = []
for label, n_pick in desired_counts.items():
    if n_pick <= 0:
        continue

    # (가) 해당 라벨에 속한 행들만 필터
    subset = dlg_final[dlg_final["최종라벨"] == label]

    # (나) subset 에서 n_pick개를 랜덤 추출 (replace=False → 중복 X, random_state 지정 → 재현 가능)
    sampled = subset.sample(n=n_pick, random_state=42, replace=False)
    sampled_list.append(sampled)
```

```python
third_sampled = pd.concat(sampled_list, ignore_index=True)
print("\n▶ 세 번째 풀에서 샘플링된 최종 데이터 개수(라벨별):")
print(third_sampled["최종라벨"].value_counts())
```

```python
# ──────────────────────────────────────────────────
# 1) third_sampled: "발화"→"text", "최종라벨"→"label"
# ──────────────────────────────────────────────────
third_renamed = third_sampled.rename(
    columns={
        "발화": "text",
        "최종라벨": "label"
    }
)[["text", "label"]]
# ▶ now third_renamed.columns == ["text", "label"]


# ──────────────────────────────────────────────────
# 2) final_df: 이미 ["text", "label"] 로 되어 있다면, 
#    해당 두 컬럼만 가져오기
# ──────────────────────────────────────────────────
# 혹시 final_df에 다른 컬럼이 섞여 있다면, 변수명 확인 후 아래처럼 두 가지만 골라냅니다.
final_renamed = final_df[["text", "label"]].copy()
# ▶ final_renamed.columns == ["text", "label"]


# ──────────────────────────────────────────────────
# 3) keti_single: "text"는 그대로, "new_label"→"label"
# ──────────────────────────────────────────────────
keti_new = keti_single[["text", "new_label"]]
keti_renamed = keti_new.rename(
    columns={
        "new_label": "label"
    }
)[["text", "label"]]
# ▶ keti_renamed.columns == ["text", "label"]

```

```python
# 4) 세 DataFrame을 리스트로 묶어서 concat
merged = pd.concat(
    [third_renamed, final_renamed, keti_renamed],
    axis=0,            # 0: 세로로 이어붙이기 (행 추가)
    ignore_index=True  # 인덱스를 0,1,2,... 순번으로 재부여
)

# 5) 결과 확인 (원하는 만큼 출력)
print("▶ 합쳐진 DataFrame 크기:", merged.shape)
print("▶ 앞부분 미리보기:")
print(merged.head(5).to_string(index=False))
print("\n▶ label별 개수 분포:")
print(merged["label"].value_counts())
```
